# Introduction to NumPy

This notebook shows basic NumPy usage, array creation, element-wise operations, and a simple performance comparison with Python lists.

```python
import numpy as np
```

## Creating NumPy Arrays

Here we create both a 1D array and a 2D array using `np.array()`.

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D array: ", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array: ", arr_2d)
```

```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2)
```

## Python List vs NumPy Array

Python lists repeat their elements when multiplied, while NumPy arrays perform element-wise multiplication.

```python
py_list = [1, 2, 3]
print("Python list multiplication ", py_list * 2)

np_array = np.array([1, 2, 3])  # element wise multiplication
print("Python array multiplication ", np_array * 2)

```

## Performance Comparison

This example compares how long it takes Python lists and NumPy arrays to perform similar multiplication operations on large data.

```python
import numpy as np
import time

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time:  ", time.time() - start)

start = time.time()
np_array = np.arange(1000000) * 2
print("\n NumPy operation time: ", time.time() - start)
```
